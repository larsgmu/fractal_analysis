### Divider Method
The divider method is the first step in the fractal analysis. This notebook is just to illustrate how it looks under the hood.

```python
# Test data, basic sine wave
A = pd.Series([np.sin(x/50) for x in range(1000)]) 
x = [i for i in range(len(A))]

# Run divider method with divider width=200
divider_width = 200
dividers = calculate_total_length(x, A, divider_width)       


# Print and plot the output
print(f"Divider width: {dividers[0]}\nPolylength: {dividers[1]}")

plt.figure(figsize=(16,9))

# Plot input signal
plt.plot(x,A)                                            

# Draw the segment lines
for b, m, start, end in dividers[2]: 
    y = generate_line(b, m, start, end)
    x_axis = [x for x in np.arange(start, end,0.01)]
    plt.plot(x_axis, y, linewidth=2)
plt.xticks(np.arange(0, len(x)+1, 50))
plt.grid()
plt.show()    
```
